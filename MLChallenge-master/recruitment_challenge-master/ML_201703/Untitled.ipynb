```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
le.fit(target.values)
target_mod=le.transform(target)
target_m=pd.DataFrame(target_mod)  
target_m.columns=['c295']
#print(target_m['c295'].value_counts(),target['c295'].value_counts())
y=target_m
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import scale
min_max=MinMaxScaler()
```

```python
numeric_scaled=scale(numeric_columns)

integer_scaled=min_max.fit_transform(integer_columns)

binary_scaled=binary_columns.as_matrix()
```

```python
import numpy as np
```

```python
numeric_scaled=np.array(numeric_scaled)
integer_scaled=np.array(integer_scaled)
binary_scaled= np.array(binary_scaled)
```

```python
print(numeric_scaled.shape,integer_scaled.shape,binary_scaled.shape)
x=np.concatenate((numeric_scaled,integer_scaled,binary_scaled), axis=1)
x.shape
```

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn import cross_validation
#Simple K-Fold cross validation. 10 folds.
X_train
#cv = cross_validation.KFold(len(X_train), n_folds=10, indices=False)
```

```python
model = MLPClassifier(hidden_layer_sizes=(100, 5), 
              activation='relu', 
              solver='adam',#because it consists of both decay and momentum 
              alpha=0.0001, 
              batch_size='auto', 
              learning_rate_init=0.001, 
              power_t=0.5, 
              max_iter=200, 
              shuffle=True, 
              random_state=None, 
              tol=0.0001, 
              verbose=False, 
              warm_start=False, 
              momentum=0.9, 
              nesterovs_momentum=True, 
              early_stopping=False, 
              validation_fraction=0.1, 
              beta_1=0.9, 
              beta_2=0.999, 
              epsilon=1e-08)
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
